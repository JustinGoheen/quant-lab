```python
import pandas as pd
import plotly.graph_objects as go
```

```python
data = pd.read_parquet("../data")
data.reset_index(inplace=True)
data.set_index("timestamp", inplace=True)
```

```python
f"the start date is {data.index[0].date()} and the end date is {data.index[-1].date()}"
```

```python
fig = go.Figure(data=[go.Candlestick(x=data.index,
                open=data['open'],
                high=data['high'],
                low=data['low'],
                close=data['close'])])

fig.update_layout(title=data["symbol"].iloc[0].upper(), width=1200)

fig.show()
```
